```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/coffee-bean-sales-dataset/DatasetForCoffeeSales2.csv")
```

```python
df.columns
```

```python
# Calculate total sales by city
total_sales_by_city = df.groupby('City')['Sales Amount'].sum().reset_index()
total_sales_by_city = total_sales_by_city.sort_values(by='Sales Amount', ascending=False)
```

```python
total_sales_by_city
```

```python
# Calculate total sales by product
total_sales_by_product = df.groupby('Product')['Sales Amount'].sum().reset_index()
top_products = total_sales_by_product.sort_values(by='Sales Amount', ascending=False).head(5)
```

```python
total_sales_by_product
```

```python
top_products
```

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate sales by date
sales_by_date = df.groupby('Date')['Sales Amount'].sum().reset_index()
sales_by_date.columns = ['ds', 'y']

# Initialize and fit the Prophet model
from prophet import Prophet
model = Prophet()
model.fit(sales_by_date)

# Create future dataframe for predictions
future = model.make_future_dataframe(periods=365)  # Predict for the next 365 days

# Make predictions
forecast = model.predict(future)


fig = model.plot(forecast)
```

```python
# Calculate average purchase amount per customer
avg_purchase_per_customer = df.groupby('Customer_ID')['Sales Amount'].mean().reset_index()
avg_purchase_per_customer.columns = ['Customer_ID', 'Average_Purchase_Amount']
```

```python
avg_purchase_per_customer
```

```python
model = Prophet(seasonality_mode='multiplicative')
model.fit(sales_by_date)

future = model.make_future_dataframe(periods=365)  

forecast = model.predict(future)
fig = model.plot_components(forecast)
```

```python
# Calculate total sales
total_sales = df['Sales Amount'].sum()

# Calculate sales contribution by city
sales_contribution_by_city = df.groupby('City')['Sales Amount'].sum().reset_index()
sales_contribution_by_city['Percentage'] = (sales_contribution_by_city['Sales Amount'] / total_sales) * 100
```

```python
# Calculate correlation between unit price and quantity sold
correlation = df[['Unit Price', 'Quantity']].corr()
```

```python
# Identify repeat customers
repeat_customers = df['Customer_ID'].value_counts().reset_index()
repeat_customers.columns = ['Customer_ID', 'Purchase_Count']
repeat_customers = repeat_customers[repeat_customers['Purchase_Count'] > 1]

retention_rate = len(repeat_customers) / len(df['Customer_ID'].unique())
```
