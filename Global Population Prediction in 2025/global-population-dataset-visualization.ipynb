```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/global-population-dataset-20142024/population_by_year_2014_2024.csv')
```

```python
df.columns
```

```python
years = [int(col) for col in df.columns if col.isdigit()]
countries = df['Country']
populations = df.iloc[:, 1:].T
```

```python
populations
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 8))
for country in countries:
    plt.plot(years, df.loc[df['Country'] == country].iloc[:, 1:].values.flatten(), label=country)

plt.title("Population Trends by Country (2014-2024)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

# Prediction

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
country_index = 0
country_name = df.iloc[country_index, 0]
```

```python
X = np.array(years).reshape(-1, 1) 
y = populations.iloc[:, country_index].values
```

```python
X_train, X_test = X[:-1], X[-1].reshape(1, -1)  # Training: 2014-2023, Testing: 2024
y_train, y_test = y[:-1], y[-1]
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)  # Predictions on training set
y_pred_test = model.predict(X_test)  # Predict for 2024 (test year)
predicted_2025 = model.predict([[2025]])  # Predict for 2025
```

```python
predicted_2025
```

```python
predicted_2025_int = int(predicted_2025[0])
```

```python
predicted_2025_int
```

```python
# Calculate accuracy metrics
mae = mean_absolute_error(y_train, y_pred_train)
mse = mean_squared_error(y_train, y_pred_train)
r2 = r2_score(y_train, y_pred_train)
```

```python
print(f"Model Performance for {country_name}:")
print(f"Mean Absolute Error (MAE): {mae:,.0f} ({(mae / y_train.mean()) * 100:.2f}%)")
print(f"Mean Squared Error (MSE): {mse:,.0f} ({(np.sqrt(mse) / y_train.mean()) * 100:.2f}%)")
print(f"R² Score: {r2:.2%}")
print(f"Predicted population for {country_name} in 2024: {y_pred_test[0]:,.0f}")
print(f"Predicted population for {country_name} in 2025: {predicted_2025[0]:,.0f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, y, label=f"Actual Population ({country_name})", marker='o')
plt.plot(years[:-1], y_pred_train, label="Model Fit (2014-2023)", linestyle='--')
plt.scatter(2024, y_pred_test, color='red', label=f"Predicted Population (2024: {y_pred_test[0]:,.0f})")
plt.scatter(2025, predicted_2025, color='purple', label=f"Predicted Population (2025: {predicted_2025[0]:,.0f})")
plt.title(f"Population Prediction for {country_name}")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True)
plt.show()
```
