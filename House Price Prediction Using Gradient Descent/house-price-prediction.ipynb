```python
# Features and target
X = df[['Square_Feet', 'Num_Bedrooms','House_Age']].values
y = df['House_Price'].values
```

```python
X
```

```python
y
```

```python
X_mean, X_std = X.mean(axis=0), X.std(axis=0)
X = (X - X_mean) / X_std
```

```python
y_mean, y_std = y.mean(), y.std()
y = (y - y_mean) / y_std
```

```python
import numpy as np
X = np.c_[np.ones(len(X)), X]
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    for _ in range(iterations):
        theta -= (alpha / m) * X.T @ (X @ theta - y)
    return theta
```

```python
theta = np.zeros(X.shape[1])
alpha = 0.1  # Adjusted learning rate
iterations = 1000
```

```python
# Train model
theta = gradient_descent(X, y, theta, alpha, iterations)
```

```python
# Prediction function
def predict(Square_Feet, Num_Bedrooms, House_Age ):
    # input using training mean/std
    Square_Feet = (Square_Feet - X_mean[0]) / X_std[0]
    Num_Bedrooms = (Num_Bedrooms - X_mean[1]) / X_std[1]
    House_Age = (House_Age - X_mean[2]) / X_std[2]

    # Compute predicted value
    normalized_price = theta[0] + theta[1] * Square_Feet + theta[2] * Num_Bedrooms + theta[3] * House_Age
    return normalized_price * y_std + y_mean
```

```python
predicted_price = predict(900, 6,40)
print(f"Predicted price for 900 sqft, 6 bedrooms, 40 years age: ${predicted_price:.2f}")
```
