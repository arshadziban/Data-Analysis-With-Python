```python
import kagglehub

# Download dataset from Kaggle
path = kagglehub.dataset_download("adilshamim8/student-depression-dataset")

print("Path to dataset files:", path)
```

```python
import os

# List all files and subfolders in the dataset directory
print("Files in the dataset directory:")
print(os.listdir(path))
```

```python

import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Use the actual file name you found
correct_file_name = "student_depression_dataset.csv" 
file_path = os.path.join(path, correct_file_name)
```

```python
# Read and show the dataset
df = pd.read_csv(file_path)
df.head()
```

# # Initial inspection

```python
df.dtypes
```

```python
df.describe(include='all')
```

```python
df.shape
```

# Data Cleaning

```python
df.isnull().sum()
```

```python
# Clean categorical columns (strip whitespace, standardize values)
df['Gender'] = df['Gender'].str.strip()
```

```python
df.head(2)
```

```python
df['Sleep Duration'] = df['Sleep Duration'].str.strip()
```

```python
df
```

```python
# Convert appropriate columns to categorical type
cat_cols = ['Gender', 'City', 'Profession', 'Sleep Duration', 'Dietary Habits', 
            'Degree', 'Have you ever had suicidal thoughts ?', 
            'Financial Stress', 'Family History of Mental Illness']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
df
```

# Demographic Analysis

```python
gender_dist = df['Gender'].value_counts(normalize=True) * 100
```

```python
gender_dist
```

```python
# Age distribution
df['Age'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

# Mental Health Indicators

```python
# Depression prevalence
depression_rate = df['Depression'].value_counts(normalize=True) * 100
```

```python
depression_rate
```

```python
# Suicidal thoughts analysis
suicidal = df['Have you ever had suicidal thoughts ?'].value_counts(normalize=True) * 100
```

```python
suicidal
```

```python
# Cross-tabulation: Depression vs. Suicidal thoughts
crosstab_result = pd.crosstab(df['Depression'], df['Have you ever had suicidal thoughts ?'], normalize='columns') * 100
```

```python
crosstab_result
```

# Academic/Work Factors

```python
# Academic pressure analysis
academic_pressure_result = df.groupby('Profession')['Academic Pressure'].mean().sort_values(ascending=False)
academic_pressure_result
```

```python
# CGPA distribution by depression status
plt.figure(figsize=(8,5))
sns.boxplot(x='Depression', y='CGPA', data=df)
plt.title('CGPA Distribution by Depression Status')
plt.xlabel('Depression Status')
plt.ylabel('CGPA')
plt.grid(True)
plt.show()
```

# Lifestyle Factors

```python
# Sleep duration analysis
sleep_depression = pd.crosstab(df['Sleep Duration'], df['Depression'], normalize='index') * 100
```

```python
sleep_depression
```

```python
# Dietary habits analysis
diet_depression = pd.crosstab(df['Dietary Habits'], df['Depression'], normalize='index') * 100
diet_depression
```

# Correlation Analysis

```python
# Select numerical columns for correlation
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[num_cols].corr()
```

```python
corr_matrix
```

```python
# Visualize correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Focus on depression correlations
print("\nCorrelation with Depression:\n", corr_matrix['Depression'].sort_values(ascending=False))
```

# Comprehensive Demographic Analysis

```python
# Depression by Gender
fig, ax = plt.subplots(figsize=(7, 6))
sns.countplot(data=df, x='Gender', hue='Depression', ax=ax)
ax.set_title('Depression by Gender')
plt.tight_layout()
plt.show()
```
