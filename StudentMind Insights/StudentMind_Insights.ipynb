```python
import kagglehub

# Download dataset from Kaggle
path = kagglehub.dataset_download("adilshamim8/student-depression-dataset")

print("Path to dataset files:", path)
```

```python
import os

# List all files and subfolders in the dataset directory
print("Files in the dataset directory:")
print(os.listdir(path))
```

```python

import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Use the actual file name you found
correct_file_name = "student_depression_dataset.csv" 
file_path = os.path.join(path, correct_file_name)
```

```python
# Read and show the dataset
df = pd.read_csv(file_path)
df.head()
```

# Data Cleaning

```python
df.isnull().sum()
```

```python
# Clean categorical columns (strip whitespace, standardize values)
df['Gender'] = df['Gender'].str.strip()
```

```python
df.head(2)
```

```python
df['Sleep Duration'] = df['Sleep Duration'].str.strip()
```

```python
df
```

```python
# Convert appropriate columns to categorical type
cat_cols = ['Gender', 'City', 'Profession', 'Sleep Duration', 'Dietary Habits', 
            'Degree', 'Have you ever had suicidal thoughts ?', 
            'Financial Stress', 'Family History of Mental Illness']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
df
```

# Demographic Analysis

```python
gender_dist = df['Gender'].value_counts(normalize=True) * 100
```

```python
gender_dist
```

```python
# Age distribution
df['Age'].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```
