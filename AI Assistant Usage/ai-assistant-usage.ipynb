```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/ai-assistant-usage-in-student-life-synthetic/ai_assistant_usage_student_life.csv")
```

```python
#What is the distribution of student academic levels using AI?
sns.countplot(x='StudentLevel', data=df)
plt.title('Student Academic Level Distribution')
plt.xlabel('Student Level')
plt.ylabel('Session Count')
plt.show()
```

```python
# Which disciplines use AI tools the most?
plt.figure(figsize=(10,5))
sns.countplot(y='Discipline', data=df, order=df['Discipline'].value_counts().index)
plt.title('AI Usage by Discipline')
plt.xlabel('Session Count')
plt.ylabel('Discipline')
plt.show()
```

```python
# Question 3: What types of tasks are students using AI for?
sns.countplot(x='TaskType', data=df)
plt.title('AI Task Type Distribution')
plt.xlabel('Task Type')
plt.ylabel('Count')
plt.show()
```

```python
# Question 4: How does session length vary by student level?
sns.boxplot(x='StudentLevel', y='SessionLengthMin', data=df)
plt.title('Session Length by Student Level')
plt.xlabel('Student Level')
plt.ylabel('Session Length (min)')
plt.show()
```

```python
# Pie chart for academic level distribution
student_counts = df['StudentLevel'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(student_counts, labels=student_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Student Academic Level Distribution')
plt.show()
```

```python
# Point plot for satisfaction by AI reuse
plt.figure(figsize=(6,4))
sns.pointplot(x='UsedAgain', y='SatisfactionRating', data=df, capsize=0.1)
plt.title('Average Satisfaction by AI Reuse')
plt.xlabel('Used AI Again')
plt.ylabel('Average Satisfaction')
plt.show()
```

```python
# Clustered bar plot for assignment completions
completion = df[df['FinalOutcome'] == 'Assignment Completed']
comp_counts = completion.groupby(['StudentLevel', 'Discipline']).size().unstack().fillna(0)
comp_counts.plot(kind='bar', figsize=(12,6))
plt.title('Assignment Completions by Student Level and Discipline')
plt.xlabel('Student Level')
plt.ylabel('Assignment Completion Count')
plt.legend(title='Discipline', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
