```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set Seaborn style
sns.set_theme(style="darkgrid")
```

```python
# Screen Time Trend Over Time
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Date", y="Screen_Time (mins)", marker="o", color="b")
plt.title("Daily Screen Time Trend")
plt.xlabel("Date")
plt.ylabel("Screen Time (mins)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Battery Level vs. Screen Time
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Battery_Level (%)", y="Screen_Time (mins)", hue="Network_Type", palette="coolwarm")
plt.title("Battery Level vs. Screen Time")
plt.xlabel("Battery Level (%)")
plt.ylabel("Screen Time (mins)")
plt.show()
```

```python
# Network Type vs. Data Usage
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x="Network_Type", y="Data_Usage (MB)", palette="Set2")
plt.title("Data Usage by Network Type")
plt.xlabel("Network Type")
plt.ylabel("Data Usage (MB)")
plt.show()
```

```python
# Most Used Apps
plt.figure(figsize=(7, 5))
sns.countplot(data=df, y="App_Usage (Top App)", order=df["App_Usage (Top App)"].value_counts().index, palette="pastel")
plt.title("Most Frequently Used Apps")
plt.xlabel("Count")
plt.ylabel("App")
plt.show()
```
